```python
data = pd.read_csv('FD-PASS_data_NatComm.csv')
```

```python
data.insert(data.shape[1], 'TreatmentSuccess', 0)
data.TreatmentSuccess[data['diff-aneuSavTavVel'] >= 35] = 1 
#Aneurysm cured or not cured
```

```python
#correlation
corr_matrix=abs(data.corr())
corr_matrix["diff-aneuSavTavVel"].sort_values(ascending=False)
```

```python
#standardization
std = StandardScaler()
data['Age'] = std.fit_transform(data['Age'].values.reshape(-1,1)) 
data['MaxDiam'] = std.fit_transform(data['MaxDiam'].values.reshape(-1,1)) 
data['Depth'] = std.fit_transform(data['Depth'].values.reshape(-1,1)) 
data['NeckWidth'] = std.fit_transform(data['NeckWidth'].values.reshape(-1,1)) 
# data['Aspect'] = std.fit_transform(data['Aspect'].values.reshape(-1,1)) 
data['SacVolume'] = std.fit_transform(data['SacVolume'].values.reshape(-1,1)) 
data['SacSurface'] = std.fit_transform(data['SacSurface'].values.reshape(-1,1)) 
data['NeckSurface'] = std.fit_transform(data['NeckSurface'].values.reshape(-1,1)) 
# data['SideBranchDiam'] = std.fit_transform(data['SideBranchDiam'].values.reshape(-1,1)) 
# data['pre-aneuMaxTavVel'] = std.fit_transform(data['pre-aneuMaxTavVel'].values.reshape(-1,1)) 
# data['pre-aneuMaxSysVel'] = std.fit_transform(data['pre-aneuMaxSysVel'].values.reshape(-1,1)) 
# data['pre-aneuSavSysVel'] = std.fit_transform(data['pre-aneuSavSysVel'].values.reshape(-1,1)) 
# data['pre-aneuMaxDiaVel'] = std.fit_transform(data['pre-aneuMaxDiaVel'].values.reshape(-1,1)) 
# data['pre-neckMaxTavVel'] = std.fit_transform(data['pre-neckMaxTavVel'].values.reshape(-1,1)) 
data['pre-neckInflowTav'] = std.fit_transform(data['pre-neckInflowTav'].values.reshape(-1,1)) 
# data['pre-neckMaxSysVel'] = std.fit_transform(data['pre-neckMaxSysVel'].values.reshape(-1,1)) 
data['pre-neckInflowSys'] = std.fit_transform(data['pre-neckInflowSys'].values.reshape(-1,1)) 
# data['pre-neckMaxDiaVel'] = std.fit_transform(data['pre-neckMaxDiaVel'].values.reshape(-1,1)) 
data['pre-neckInflowDia'] = std.fit_transform(data['pre-neckInflowDia'].values.reshape(-1,1)) 
# data['diff-aneuSavTavVel'] = std.fit_transform(data['diff-aneuSavTavVel'].values.reshape(-1,1)) 
```

```python
#Stratifiedsampling
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=0)
for train_index, test_index in split.split(data, data['TreatmentSuccess']):
    trainset = data.loc[train_index]
    testset = data.loc[test_index]
```

```python
#Trainset and testset
X = data.drop(['TreatmentSuccess','Case ID','Cerebral','Gender','diff-aneuSavTavVel',
                      'pre-aneuSavTavVel','pre-neckInflowTav','pre-neckMaxSysVel','pre-aneuSavSysVel','pre-neckInflowSys','pre-aneuSavDiaVel','Case ID','Cerebral','Gender','Age','Gender','MaxDiam','Depth','NeckWidth','Aspect','SacVolume','SacSurface','NeckSurface','Non-Spherecity'
                      ],axis=1)
y = data['TreatmentSuccess']


train_X = trainset.drop(['TreatmentSuccess','Case ID','Cerebral','Gender','diff-aneuSavTavVel',
                        'pre-aneuSavTavVel','pre-neckInflowTav','pre-neckMaxSysVel','pre-aneuSavSysVel','pre-neckInflowSys','pre-aneuSavDiaVel','Case ID','Cerebral','Gender','Age','Gender','MaxDiam','Depth','NeckWidth','Aspect','SacVolume','SacSurface','NeckSurface','Non-Spherecity'
                        ],axis=1)
test_X = testset.drop(['TreatmentSuccess','Case ID','Cerebral','Gender','diff-aneuSavTavVel',
                      'pre-aneuSavTavVel','pre-neckInflowTav','pre-neckMaxSysVel','pre-aneuSavSysVel','pre-neckInflowSys','pre-aneuSavDiaVel','Case ID','Cerebral','Gender','Age','Gender','MaxDiam','Depth','NeckWidth','Aspect','SacVolume','SacSurface','NeckSurface','Non-Spherecity'
                      ],axis=1)
# train_y = trainset['diff-aneuSavTavVel']
# test_y = testset['diff-aneuSavTavVel']


train_y = trainset['TreatmentSuccess']
test_y = testset['TreatmentSuccess']
```

```python
train_X.shape,train_y.shape,test_X.shape,test_y.shape
```

```python
#XGB regression
clf = XGBClassifier()
# Train the XGBoost model on the training set
clf.fit(train_X, train_y)

# prediction
train_predict = clf.predict(train_X)
test_predict = clf.predict(test_X)
from sklearn import metrics

# Evaluate model effectiveness using accuracy
print('The accuracy of the XGBClassifier is:',metrics.accuracy_score(train_y,train_predict))
print('The accuracy of the XGBClassifier is:',metrics.accuracy_score(test_y,test_predict))

# Review the confusion matrix
confusion_matrix_result = metrics.confusion_matrix(test_predict,test_y)
print('The confusion matrix result:\n',confusion_matrix_result)

# Visualize results with heat maps
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix_result, annot=True, cmap='Blues')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```
