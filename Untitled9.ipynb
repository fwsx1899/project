```python
#Stratifiedsampling
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=0)
for train_index, test_index in split.split(data, data['TreatmentSuccess']):
    trainset = data.loc[train_index]
    testset = data.loc[test_index]
```

```python
#Trainset and testset
X = data.drop(['TreatmentSuccess','Case ID','Cerebral','Gender','diff-aneuSavTavVel',
                      'pre-aneuSavTavVel','pre-neckInflowTav','pre-neckMaxSysVel','pre-aneuSavSysVel','pre-neckInflowSys','pre-aneuSavDiaVel','Case ID','Cerebral','Gender','Age','Gender','MaxDiam','Depth','NeckWidth','Aspect','SacVolume','SacSurface','NeckSurface','Non-Spherecity'
                      ],axis=1)
y = data['TreatmentSuccess']


train_X = trainset.drop(['TreatmentSuccess','Case ID','Cerebral','Gender','diff-aneuSavTavVel',
                        'pre-aneuSavTavVel','pre-neckInflowTav','pre-neckMaxSysVel','pre-aneuSavSysVel','pre-neckInflowSys','pre-aneuSavDiaVel','Case ID','Cerebral','Gender','Age','Gender','MaxDiam','Depth','NeckWidth','Aspect','SacVolume','SacSurface','NeckSurface','Non-Spherecity'
                        ],axis=1)
test_X = testset.drop(['TreatmentSuccess','Case ID','Cerebral','Gender','diff-aneuSavTavVel',
                      'pre-aneuSavTavVel','pre-neckInflowTav','pre-neckMaxSysVel','pre-aneuSavSysVel','pre-neckInflowSys','pre-aneuSavDiaVel','Case ID','Cerebral','Gender','Age','Gender','MaxDiam','Depth','NeckWidth','Aspect','SacVolume','SacSurface','NeckSurface','Non-Spherecity'
                      ],axis=1)
# train_y = trainset['diff-aneuSavTavVel']
# test_y = testset['diff-aneuSavTavVel']


train_y = trainset['TreatmentSuccess']
test_y = testset['TreatmentSuccess']
```

```python
#XGB regression
clf = LogisticRegression() 
# clf = RandomForestClassifier()
# clf = XGBClassifier()
# Train the XGBoost model on the training set
clf.fit(train_X, train_y)

# prediction
train_predict = clf.predict(train_X)
test_predict = clf.predict(test_X)
from sklearn import metrics

# Evaluate model effectiveness using accuracy
print('The accuracy of the Logistic Regression is:',metrics.accuracy_score(train_y,train_predict))
print('The accuracy of the Logistic Regression is:',metrics.accuracy_score(test_y,test_predict))

# Review the confusion matrix
confusion_matrix_result = metrics.confusion_matrix(test_predict,test_y)
print('The confusion matrix result:\n',confusion_matrix_result)

# Visualize results with heat maps
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix_result, annot=True, cmap='Blues')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.show()
```

```python

lrl1 = LogisticRegression(penalty="l1", solver = "liblinear", C = 0.5, max_iter=1000)
lrl2 = LogisticRegression(penalty="l2", solver = "liblinear", C = 0.5, max_iter=1000)
 
lrl1 = lrl1.fit(X,y)
lrl2 = lrl2.fit(X,y)
 
(lrl1.coef_!=0).sum(axis=1) #l1
(lrl2.coef_!=0).sum(axis=1) #l2
 
#Debug parameters
l1test=[]
l2test=[]
l1 = []
l2 = []
 
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,y, test_size=0.3, random_state=42)
for i in np.linspace(0.05, 1, 19):
    lrl1 = LogisticRegression(penalty="l1", solver = "liblinear", C = i, max_iter=1000)
    lrl2 = LogisticRegression(penalty="l2", solver = "liblinear", C = i, max_iter=1000)
    
    lrl1 = lrl1.fit(Xtrain, Ytrain)
    l1.append(accuracy_score(lrl1.predict(Xtrain), Ytrain))
    l1test.append(accuracy_score(lrl1.predict(Xtest), Ytest))
    
    lrl2 = lrl2.fit(Xtrain, Ytrain)
    l2.append(accuracy_score(lrl2.predict(Xtrain), Ytrain))
    l2test.append(accuracy_score(lrl2.predict(Xtest), Ytest))
    
graph = [l1, l2, l1test, l2test]
color = ["green","black","lightgreen","gray"]
label = ["L1","L2","L1test", "L2test"]
 
plt.figure(figsize=(6,6))
for i in range(len(graph)):
    plt.plot(np.linspace(0.05,1, 19),graph[i], color[i], label = label[i])
plt.legend(loc=4) #loc = 4
plt.show()
```
